```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the Netflix dataset
df = pd.read_csv('netflix1.csv')
```

```python
# 1. Line Graph – Number of releases over the years
release_counts = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(20, 5))
plt.plot(release_counts.index, release_counts.values, marker='o')
plt.title('Number of Releases Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# 2. Bar Plot – Movies vs TV Shows
type_counts = df['type'].value_counts() # kati wota chai movies ani kati wota series (tv)
plt.figure(figsize=(6, 4))
plt.bar(type_counts.index, type_counts.values, color=['skyblue', 'salmon'])
plt.title('Number of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# 3. Histogram – Distribution of Release Years
plt.figure(figsize=(10, 5))
plt.hist(df['release_year'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.show()
```

```python
# 5. Pie Chart – Top 6 Content Ratings
rating_counts = df['rating'].value_counts().head(6)
plt.figure(figsize=(7, 7))
plt.pie(rating_counts.values, labels=rating_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 6 Content Ratings on Netflix')
plt.tight_layout()
plt.show()
```

```python
# 6. Extra Bar Plot – Top 10 Countries by Title Count
country_counts = df['country'].value_counts().head(10)
plt.figure(figsize=(10, 5))
plt.bar(country_counts.index, country_counts.values, color='orange')
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
