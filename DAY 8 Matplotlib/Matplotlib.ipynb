## Warm-Up: What is Data Visualization?

Imagine you have exam scores of 100 students in your notebook. Instead of reading every single score, wouldn’t it be easier to just look at a chart and instantly understand how well the class performed?

That’s **data visualization** — turning data into visuals like charts and graphs.

- Why is it important? Because it helps us understand patterns, trends, and insights easily.
- Common charts: Line plots, bar charts, scatter plots, and histograms.

Let’s first make sure Matplotlib is installed and check its version.

```python
import matplotlib.pyplot as plt
```

## Introduction to Matplotlib

**Matplotlib** is a library in Python that lets us create plots easily.

Within Matplotlib, the **`pyplot`** module is like a toolbox with simple commands to create charts — similar to how buttons on a microwave make it easy to cook without reading a manual.

Let’s create our very first plot!

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.show()
```

This creates a **line plot** — think of it like connecting dots on a page to show growth or trends.

### Real-Life Analogy:
Imagine tracking your savings every month. This kind of graph helps you see if you’re saving more over time!

```python
Month = ['Baisakh', 'Jestha', 'Asar','Shrawan','Bhadra']
Saving = [2000, 1000, 3000, 1500, 2500]

plt.plot(Month, Saving)
plt.show()
```

## Line Plot with Customization
Let’s now add labels, a title, and a legend to our line chart — just like adding labels to a map.


```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, label="Line Plot")
plt.xlabel("Number from 1 to 5")
plt.ylabel("Square of the number")
plt.title("Simple Line Plot")
plt.legend()
plt.show()
```

## Bar Plot
Bar plots are like comparing the height of buildings. Each bar represents a category.


```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.bar(categories, values, color='red')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot Example")
plt.show()
```

## Scatter Plot
Scatter plots are like throwing paint on a canvas — each dot represents a pair of values.

**Real-Life Analogy:** Plotting height vs. weight of students to look for trends.

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.scatter(x, y, color='red')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter Plot Example")
plt.show()
```

## Histogram
Histograms are great for showing how data is spread out.

**Real-Life Example:** Grouping test scores to see how many students scored in each range.

```python
data = [12, 12,15, 14, 10, 10, 15, 14, 14, 14,17, 18, 13, 13, 12, 13]

plt.hist(data, bins=5, color='green')
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

## Customizing Plots
We can change colors, styles, and markers — think of it like decorating a cake to make it attractive.

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, color='purple', linestyle='--', marker='o', label="Customized Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Customized Line Plot")
plt.legend()
plt.show()
```

## Activity: Plotting a Dataset
Now let’s use a sample dataset to create a bar chart.

**Real-Life Example:** You run a lemonade stand and track sales in different areas.


```python
import pandas as pd

data = {'Category': ['A', 'B', 'C', 'D'],
         'Frequency': [10, 25, 15, 30]
}
df = pd.DataFrame(data)

plt.bar(df['Category'], df['Frequency'])
plt.xlabel("Category")
plt.ylabel("Values")
plt.title("Category vs Values")
plt.show()
```
