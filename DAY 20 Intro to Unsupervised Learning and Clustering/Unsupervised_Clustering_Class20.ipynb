## 1. Introduction to Unsupervised Learning
Unsupervised learning is a type of machine learning where we don't provide labels to the learning algorithm. Instead, it tries to discover patterns, structures, or groupings in the data on its own.

### Why Use Unsupervised Learning?
- You have data but no predefined labels.
- Useful in real-world cases like:
  - Customer segmentation
  - Anomaly detection
  - Dimensionality reduction
  - Association mining

### Key Techniques:
- **Clustering** ‚Äì Group similar data points together.
- **Dimensionality Reduction** ‚Äì Reduce the number of input features while preserving essential information.



## 2. Introduction to Clustering and Its Applications

Clustering is the process of grouping similar data points based on their features. It helps uncover hidden patterns.

### Real-Life Applications:
- **Customer Segmentation**: Identify types of customers based on behavior.
- **Market Basket Analysis**: Identify sets of items frequently bought together.
- **Image Segmentation**: Divide images into regions.
- **Anomaly Detection**: Find abnormal behavior or outliers.

### Types of Clustering:
- **Partitional**: Splits into non-overlapping groups. Example: K-Means.
- **Hierarchical**: Builds a tree of clusters. Example: Agglomerative Clustering.
- **Density-Based**: Finds clusters of high density. Example: DBSCAN.



## 3. Clustering Techniques

### 1. Partitional Clustering ‚Äì K-Means

K-Means tries to divide data into `k` groups (clusters). It works by placing `k` centroids and assigning data points to the nearest one.

#### Steps:
1. Initialize `k` centroids randomly.
2. Assign each point to the closest centroid.
3. Recalculate centroids (mean of points in each cluster).
4. Repeat steps 2-3 until centroids don‚Äôt move.

#### Pros:
- Efficient and fast
- Good for large datasets

#### Cons:
- You must decide `k` in advance
- Sensitive to initial centroids


### üîç Example: K-Means on Iris Dataset

```python

from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load dataset
data = load_iris()
X = data.data

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Visualize
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```


**Explanation:**  
We use the Iris dataset (without labels) and apply KMeans to divide it into 3 clusters. The color shows which cluster each data point belongs to.



### 2. Hierarchical Clustering

This method creates a tree of clusters.

- **Agglomerative (Bottom-Up)**: Start with each data point as its own cluster, then merge.
- **Divisive (Top-Down)**: Start with one big cluster, then split.

#### Dendrogram:
A tree diagram to visualize the merging of clusters.

#### Pros:
- No need to specify number of clusters
- Useful for hierarchy

#### Cons:
- Slow on large datasets


```python

from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Agglomerative Clustering
agg_clust = AgglomerativeClustering(n_clusters=3)
y_agg = agg_clust.fit_predict(X)

# Dendrogram
linked = linkage(X, 'single')
dendrogram(linked)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```


**Explanation:**  
Agglomerative Clustering merges the closest data points one by one. A dendrogram shows how clusters are merged. You can "cut" the tree at different heights to choose number of clusters.



### 3. Density-Based Clustering ‚Äì DBSCAN

This method finds clusters by looking for high-density areas.

#### Parameters:
- **Epsilon (Œµ)**: Max distance between two samples to be considered neighbors.
- **MinPts**: Minimum number of points to form a dense region.

#### Pros:
- Finds clusters of arbitrary shapes
- Handles outliers well

#### Cons:
- Sensitive to Œµ and MinPts
- Struggles with clusters of different densities


```python

from sklearn.cluster import DBSCAN

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Visualize
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis')
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```


**Explanation:**  
DBSCAN detects clusters by grouping points that are close together. Points not in any cluster are labeled as `-1` (outliers).
